```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from urllib.request import urlopen
import numpy as np
from matplotlib import pyplot as plt
```

# nirf_2021


```python
def nirf_2021():
  url ="https://www.nirfindia.org/Rankings/2021/EngineeringRanking.html"
  html = urlopen(url)
  soup = bs(html,'html.parser')
  tables = soup.find_all("table",{"id":"tbl_overall"})
  th = tables[0].find_all("th")
  tr = tables[0].find_all("tr")
  column = [(i.text) for i in th[:11]]
  column_name = column[:2]+column[6:11]+column[2:6]
  table_data = []
  temp_name = []

  for j in range(1,len(tr)):
    temp_data = []
    for k,i in enumerate(tr[j].find_all("td")):
      if k == 1:
        if len(i.text.split("More")) != 1:
          temp_name.append(i.text.split("More")[0])
          temp_data.append(i.text.split("More")[0])

      else:
        temp_data.append(i.text)
    if len(temp_data) != 0 and len(temp_data) != 4 :
        table_data.append(temp_data)

  table_data = table_data[:100]
  nirf_2021 = pd.DataFrame(table_data,columns=column_name)
  nirf_2021.to_csv("/content/drive/MyDrive/NIRF/nirf_2021",index=False)
  print("NIRF Ranking of 2021")
  display(nirf_2021)
```

```python
ranking_2021 = pd.read_csv('/content/drive/MyDrive/NIRF/nirf_2021').iloc[:20]
ranking_2022 = pd.read_csv('/content/drive/MyDrive/NIRF/nirf_2022')
ranking_2023 = pd.read_csv('/content/drive/MyDrive/NIRF/nirf_2023')
ranking_2024 = pd.read_csv('/content/drive/MyDrive/NIRF/nirf_2024')
```

```python
def merge_and_save(metric, file_name):
    combined_table = ((ranking_2021.merge(ranking_2022, on=['Institute ID']))
                                 .merge(ranking_2023, on='Institute ID'))

    temp_df = combined_table.loc[:, ['Institute ID', 'Name_x', f'{metric}_x', f'{metric}_y', metric]]
    temp_df.columns = ['Institute ID', 'Name', f'{metric}_2021', f'{metric}_2022', f'{metric}_2023']

    combined_table = temp_df.merge(ranking_2024, on='Institute ID')
    combined_table = combined_table.loc[:, ['Institute ID', 'Name_x', f'{metric}_2021', f'{metric}_2022', f'{metric}_2023', metric]]
    combined_table.columns = ['Institute ID', 'Name', f'{metric}_2021', f'{metric}_2022', f'{metric}_2023', f'{metric}_2024']

    combined_table.to_csv(f'/content/drive/MyDrive/NIRF/{file_name}.csv', index=False)

# List of metrics and corresponding filenames
metrics = {
    'Rank': 'rank',
    'TLR (100)': 'TLR',
    'RPC (100)': 'RPC',
    'GO (100)': 'GO',
    'OI (100)': 'OI',
    'PERCEPTION (100)': 'PERCEPTION'
}

# Loop over the metrics and call the function
for metric, file_name in metrics.items():
    merge_and_save(metric, file_name)
```
